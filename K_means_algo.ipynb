```python
import numpy as np
import matplotlib.pyplot as plt
# The module name is 'sklearn', not 'Sklearn'
from sklearn.cluster import KMeans
#Corrected the numpy array to have the correct shape. Each data point is an array [feature1,feature2]
x = np.array([[15,80], [16,75],[20,90], [20,90], [50,30],[55,25]])
#Initialize the KMeans object with a consistent variable name Kmeans and fit the data.
Kmeans = KMeans(n_clusters=2, random_state=42)
Kmeans.fit(x)
centroids = Kmeans.cluster_centers_
plt.scatter(x[:,0], x[:,1], c=Kmeans.labels_, cmap='rainbow')
plt.scatter(centroids[:,0], centroids[:,1], c='black', marker='x',label='centroids')
# Corrected the typo from 'xlebel' to 'xlabel'
plt.xlabel('Annual Income($1000s)')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
